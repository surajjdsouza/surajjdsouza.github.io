```python
import nltk
from nltk.chat.util import Chat, reflections

# 1. Define the PoC Rule Set
poc_rules = [
    [
        r"(hi|hello|hey)",
        ["Hello! I am the IT Support underling. How may I be of service?"]
    ],
    [
        r"my (password|account) is (locked|broken|expired)",
        ["Have you tried visiting the self-service %1 reset portal?"]
    ],
    [
        r"my (machine|desktop|laptop) is (not working|broken|blank|not responsive)",
        ["Have you tried turning it off and on again?"]
    ],
    [
        r"the printer (isn't|is not) printing",
        ["Is it plugged in?"]
    ],
    [
        r"the printer says PC LOAD LETTER",
        ["Do you have a baseball bat nearby?"]
    ],
    [
        r"i need a new (laptop|mouse|keyboard|monitor)",
        ["Hardware requests for a new %1 require meatbag approval. Please submit form IT-9000."]
    ],
    [
        r"who created you?",
        ["I am a hardcoded proof of concept built in Python!"]
    ],
    [
        r"(quit|exit)",
        ["Shutting down for now. Goodbye!"]
    ],
    [
        r"(.*)",
        ["As a machine based lifeform, I don't understand '%1' yet. Try asking about machine related issues."]
    ]
]

# 2. Launch the PoC
def run_poc():
    print("--- PoC Chatbot Initialized ---")
    print("Type 'quit' to exit.")
    print("-" * 31)

    chatbot = Chat(poc_rules, reflections)
    chatbot.converse()

run_poc()
```
